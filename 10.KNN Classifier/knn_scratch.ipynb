# Implementing KNN from scratch

## Importing Modules

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Getting Data

```python
from sklearn import datasets

X1, y = datasets.make_blobs(n_samples=100, n_features=2, centers=2, cluster_std=1.05, random_state=50)
X2 = np.array(np.random.uniform(low=-6, high=3, size=(15,2)))
X1.shape, y.shape, X2.shape
```

## Plotting Data

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x=X1[:, 0], y=X1[:, 1], hue=y, s=50)
sns.scatterplot(x=X2[:, 0], y=X2[:, 1], color="#111823", marker="s", s=100, label="predict")
plt.xlim(-10, 5)
plt.ylim(-10, 5)
```

## Predicting Values

```python
def euclidian(x1, x2):
    return np.sqrt(np.square(x1[:, 0] - x2[0]) + np.square(x1[:, 1] - x2[1]))
```

```python
def prediction(n1, n2, *, clusters):
    pred = np.array([])
    for pts in n2:
        distance = euclidian(n1, pts)
        indices = np.argpartition(distance, clusters)[:clusters]
        maximum = [1 if np.sum(y[indices]) > (clusters / 2) else 0]
        #print(y[indices], maximum)
        pred = np.append(pred, maximum)
    
    return pred
```

```python
pred = prediction(X1, X2, clusters=3)
```

## Plotting Predictions

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x=X1[:, 0], y=X1[:, 1], hue=y, s=50)
sns.scatterplot(x=X2[:, 0], y=X2[:, 1], hue=pred, marker="s", s=100)

plt.xlim(-10, 5)
plt.ylim(-10, 5)
```
